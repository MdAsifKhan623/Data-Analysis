```python
banking=pd.read_csv('Banking.csv')
banking.head()
```

```python
banking.isna().sum()
```

```python
banking=banking.dropna()
```

```python
print(banking.shape)
```

```python
banking.isna().sum()
```

```python
bankingCategoryColumns=banking.select_dtypes(exclude=np.number).columns
```

```python
print(bankingCategoryColumns)
```

```python
banking[bankingCategoryColumns]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder=LabelEncoder()
```

```python
for i in bankingCategoryColumns:
    banking[i]=labelEncoder.fit_transform(banking[i])
```

```python
banking.head()
```

```python
banking[bankingCategoryColumns].head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoding = OneHotEncoder(sparse=False)
one_hotEncoded = one_hot_encoding.fit_transform(banking[bankingCategoryColumns])
```

```python
one_hotEncoded
```

```python
new_onehot_encoded_dataFrame = pd.DataFrame(one_hotEncoded,
                                            columns=one_hot_encoding.get_feature_names(bankingCategoryColumns))
```
