```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import chi2
```

```python
train_data_X=train_data.iloc[:,:41]
categoryCols=train_data_X.select_dtypes(exclude=np.number).columns
print(categoryCols)
```

```python
labelencoder=LabelEncoder()
for col in categoryCols:
    train_data_X[col]=labelencoder.fit_transform(train_data_X[col])
train_data_X.head()
```

```python
train_data_Y=train_data.iloc[:,-1]
Y_train=pd.DataFrame(train_data_Y)
Y_train['class'].replace({'normal':1.0,'anomaly':0.0},inplace=True)
```

```python
#Apply SelectKBest to extract 10 best features in the training dataset
bestfeatures = SelectKBest(score_func=chi2, k=10)
Y_train.head()
```

```python
fit=bestfeatures.fit(train_data_X,Y_train)
```

```python
dfscores = pd.DataFrame(fit.scores_)
```

```python
dfcolumns = pd.DataFrame(train_data_X.columns)
```

```python
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Cols','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
X = train_data_X[['src_bytes', 'dst_bytes', 'duration','dst_host_srv_count','count','dst_host_count','service','flag',
                     'dst_host_srv_serror_rate','srv_serror_rate']].copy()
```

```python
X.head()
```

```python
Y=train_data.iloc[:,-1]
```

```python
Y=Y_train
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.2, random_state=0)
x_test.head()
```

```python
from xgboost import XGBClassifier
xgbModel=XGBClassifier()

xgbModel.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred=xgbModel.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
